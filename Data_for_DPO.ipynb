```python
# ===== Step 5: setup, discover papers, preflight (Full Tier 30x2) =====
import os, json, time, shutil, random, subprocess
from pathlib import Path

REPO_DIR = Path("/content/Rosa")
DATASET_ROOT = Path("/content/drive/MyDrive/papertok_dataset_cs229")
PAPERS_ROOT = DATASET_ROOT / "papers"
RUNS_ROOT = DATASET_ROOT / "stage2_runs_full_test"   # <-- new run folder
RUNS_ROOT.mkdir(parents=True, exist_ok=True)

NUM_PAPERS = 3
NUM_CANDIDATES_PER_PAPER = 2
PER_PAPER_TIMEOUT_SEC = 1200

MANIM_GATE_TIER = "3D-LITE"
MODEL_NAME = "Qwen/Qwen2.5-Coder-7B-Instruct"
BASE_URL = "http://127.0.0.1:1234/v1"

# Qwen-only retries (no Gemini fallback for correction)
os.environ["LMSTUDIO_DISABLE_GEMINI_FALLBACK"] = "1"

assert REPO_DIR.exists(), f"Repo missing: {REPO_DIR}"
assert DATASET_ROOT.exists(), f"Dataset root missing: {DATASET_ROOT}"
assert PAPERS_ROOT.exists(), f"Papers root missing: {PAPERS_ROOT}"

# Find papers that have slides.json from stage 1
paper_dirs = sorted([
    p for p in PAPERS_ROOT.glob("paper_*")
    if (p / "outputs" / "slides.json").exists()
])

assert len(paper_dirs) > 0, f"No usable papers found under {PAPERS_ROOT}"
selected_paper_dirs = paper_dirs[:NUM_PAPERS]

print(f"Found papers: {len(paper_dirs)}")
print(f"Selected papers: {len(selected_paper_dirs)}")
for p in selected_paper_dirs[:5]:
    print(" -", p)

# Critical preflight: ensure uv environment can import manim
res = subprocess.run(
    ["uv", "run", "python", "-c", "import manim; print('manim_ok', manim.__version__)"],
    cwd=REPO_DIR,
    capture_output=True,
    text=True
)
print("Preflight rc:", res.returncode)
print("Preflight stdout:", res.stdout.strip())
print("Preflight stderr:", res.stderr.strip())
assert res.returncode == 0, "Manim import failed in uv env."
```

```python
# ===== Step 7: full tier generation (30 papers x 2 candidates) =====
run_log = []
total_jobs = len(selected_paper_dirs) * NUM_CANDIDATES_PER_PAPER
job_num = 0

for paper_src in selected_paper_dirs:
    paper_id = paper_src.name

    for cand_idx in range(1, NUM_CANDIDATES_PER_PAPER + 1):
        job_num += 1
        run_id = f"{paper_id}_cand{cand_idx}"
        run_root = RUNS_ROOT / run_id

        if run_root.exists():
            shutil.rmtree(run_root)
        shutil.copytree(paper_src, run_root)

        env = os.environ.copy()
        env["MANIM_GATE_TIER"] = MANIM_GATE_TIER
        env["LMSTUDIO_MODEL"] = MODEL_NAME
        env["LMSTUDIO_BASE_URL"] = BASE_URL
        env["LMSTUDIO_TEMPERATURE"] = str(round(random.uniform(0.2, 0.8), 2))
        env["LMSTUDIO_DISABLE_GEMINI_FALLBACK"] = "1"

        cmd = [
            "uv", "run", "python", "-m", "src.pipeline",
            str(run_root), "--from-slides", "--quality", "l"
        ]

        t0 = time.time()
        rc = None
        stderr_tail = ""

        try:
            proc = subprocess.run(
                cmd, cwd=REPO_DIR, env=env, capture_output=True, text=True, timeout=PER_PAPER_TIMEOUT_SEC
            )
            rc = proc.returncode
            stderr_tail = "\n".join(proc.stderr.splitlines()[-50:])
        except subprocess.TimeoutExpired:
            rc = 124
            stderr_tail = "timeout"

        elapsed = round(time.time() - t0, 2)
        print(f"[{job_num}/{total_jobs}] {run_id} rc={rc} t={elapsed}s")

        run_log.append({
            "paper_id": paper_id,
            "candidate_id": f"cand{cand_idx}",
            "run_id": run_id,
            "root": str(run_root),
            "rc": rc,
            "elapsed_sec": elapsed,
            "stderr_tail": stderr_tail,
        })

log_path = DATASET_ROOT / "stage2_run_log_bplus_20x2.json"
with open(log_path, "w", encoding="utf-8") as f:
    json.dump(run_log, f, ensure_ascii=False, indent=2)

print("\nSaved run log:", log_path)
print("Total runs:", len(run_log))
print("Non-zero rc:", sum(1 for x in run_log if x["rc"] != 0))
```

Build candidate index + pair pool + labeling template

```python
# ===== Step 8: build index/pair pool/label template for B+ 30x2 =====
import csv, json, itertools
from collections import defaultdict
from pathlib import Path

INDEX_OUT = DATASET_ROOT / "dpo_candidate_index_bplus_30x2.jsonl"
PAIR_OUT = DATASET_ROOT / "dpo_pair_pool_auto_bplus_30x2.jsonl"
LABEL_OUT = DATASET_ROOT / "dpo_labels_template_bplus_30x2.csv"

def parse_bool(v):
    return str(v).strip().lower() in {"1", "true", "yes"}

def parse_int(v, d=0):
    try:
        return int(v)
    except:
        return d

def parse_float(v, d=-1.0):
    try:
        return float(v)
    except:
        return d

def pick_best_row(rows):
    # compiled first, then higher attempt, then higher reward
    rows = sorted(
        rows,
        key=lambda r: (
            1 if parse_bool(r.get("compiled", "false")) else 0,
            parse_int(r.get("attempt", 0), 0),
            parse_float(r.get("reward", -1.0), -1.0),
        ),
        reverse=True,
    )
    return rows[0] if rows else None

candidate_rows = []
grouped_by_slide = defaultdict(dict)  # (paper_id, slide_id) -> {candX: best_row_rec}

# Read each run's pipeline_results.csv
for run_dir in sorted(RUNS_ROOT.glob("paper_*_cand*")):
    run_name = run_dir.name
    if "_cand" not in run_name:
        continue

    paper_id, cand_suffix = run_name.rsplit("_cand", 1)
    candidate_id = f"cand{cand_suffix}"

    pr = run_dir / "outputs" / "pipeline_results.csv"
    if not pr.exists():
        continue

    by_slide = defaultdict(list)
    with open(pr, "r", encoding="utf-8") as f:
        reader = csv.DictReader(f)
        for row in reader:
            slide_id = row.get("slide_id") or row.get("id") or ""
            if not slide_id:
                continue
            by_slide[slide_id].append(row)

    for slide_id, rows in by_slide.items():
        best = pick_best_row(rows)
        if best is None:
            continue

        rec = {
            "paper_id": paper_id,
            "candidate_id": candidate_id,
            "run_id": run_name,
            "slide_id": slide_id,
            "attempt": parse_int(best.get("attempt", 1), 1),
            "status": best.get("status", ""),
            "compiled": parse_bool(best.get("compiled", "false")),
            "compile_retries": parse_int(best.get("compile_retries", 0), 0),
            "reward": parse_float(best.get("reward", -1.0), -1.0),
            "error_bucket": best.get("error_bucket", ""),
            "gate_errors": best.get("gate_errors", ""),
            "video_file": best.get("video_file", ""),
            "manim_file": best.get("manim_file", ""),
            "prompt_user": best.get("prompt_user", ""),
            "generated_code": best.get("generated_code", ""),
            "stderr_or_error": best.get("stderr_or_error", ""),
            "root": str(run_dir),
        }
        candidate_rows.append(rec)
        grouped_by_slide[(paper_id, slide_id)][candidate_id] = rec

# Write candidate index
with open(INDEX_OUT, "w", encoding="utf-8") as f:
    for r in candidate_rows:
        f.write(json.dumps(r, ensure_ascii=False) + "\n")

def score(r):
    return (
        1 if r["compiled"] else 0,
        -int(r["compile_retries"]),
        float(r["reward"]),
    )

pairs = []
pair_num = 0

# Build all pairwise pairs among available candidates for each slide
for (paper_id, slide_id), cmap in grouped_by_slide.items():
    cand_ids = sorted(cmap.keys())  # e.g., cand1,cand2,cand3
    if len(cand_ids) < 2:
        continue

    for a_id, b_id in itertools.combinations(cand_ids, 2):
        a = cmap[a_id]
        b = cmap[b_id]
        chosen, rejected = (a, b) if score(a) >= score(b) else (b, a)

        pair_num += 1
        pairs.append({
            "pair_id": f"pair_{pair_num:06d}",
            "paper_id": paper_id,
            "slide_id": slide_id,
            "prompt": chosen["prompt_user"] or rejected["prompt_user"],
            "chosen_candidate": chosen["run_id"],
            "rejected_candidate": rejected["run_id"],
            "chosen_compiled": chosen["compiled"],
            "rejected_compiled": rejected["compiled"],
            "chosen_code": chosen["generated_code"],
            "rejected_code": rejected["generated_code"],
            "chosen_video": chosen["video_file"],
            "rejected_video": rejected["video_file"],
            "auto_rule": "compile_then_retries_then_reward",
            "label": "",
            "notes": "",
        })

# Write pair pool
with open(PAIR_OUT, "w", encoding="utf-8") as f:
    for p in pairs:
        f.write(json.dumps(p, ensure_ascii=False) + "\n")

# Write labeling template CSV
with open(LABEL_OUT, "w", encoding="utf-8", newline="") as f:
    w = csv.DictWriter(f, fieldnames=[
        "pair_id", "paper_id", "slide_id",
        "chosen_candidate", "rejected_candidate",
        "chosen_compiled", "rejected_compiled",
        "label", "notes"
    ])
    w.writeheader()
    for p in pairs:
        w.writerow({
            "pair_id": p["pair_id"],
            "paper_id": p["paper_id"],
            "slide_id": p["slide_id"],
            "chosen_candidate": p["chosen_candidate"],
            "rejected_candidate": p["rejected_candidate"],
            "chosen_compiled": p["chosen_compiled"],
            "rejected_compiled": p["rejected_compiled"],
            "label": "",
            "notes": "",
        })

print("Saved:", INDEX_OUT)
print("Saved:", PAIR_OUT)
print("Saved:", LABEL_OUT)
```

```python
# ===== Step 9: report counts =====
from pathlib import Path

INDEX_OUT = Path("/content/drive/MyDrive/papertok_dataset_cs229/dpo_candidate_index_bplus_30x2.jsonl")
PAIR_OUT = Path("/content/drive/MyDrive/papertok_dataset_cs229/dpo_pair_pool_auto_bplus_30x2.jsonl")

def count_lines(fp):
    if not fp.exists():
        return 0
    with open(fp, "r", encoding="utf-8") as f:
        return sum(1 for _ in f)

def count_compiled_true(fp):
    if not fp.exists():
        return 0
    n = 0
    with open(fp, "r", encoding="utf-8") as f:
        for line in f:
            if '"compiled": true' in line:
                n += 1
    return n

print("Indexed candidates:", count_lines(INDEX_OUT))
print("Compiled true in index:", count_compiled_true(INDEX_OUT))
print("Built auto pair pool:", count_lines(PAIR_OUT))
```

```python
# ===== Step 6.5: compile sanity + bucket breakdown (3D-LITE smoke) =====
import csv
from pathlib import Path
from collections import Counter

smoke_prefix = selected_paper_dirs[0].name + "_cand"
smoke_dirs = sorted([p for p in RUNS_ROOT.glob(f"{smoke_prefix}*") if p.is_dir()])

print("Smoke dirs:", [d.name for d in smoke_dirs])

total_rows = 0
compiled_rows = 0
bucket_counter = Counter()

for run_dir in smoke_dirs:
    fp = run_dir / "outputs" / "pipeline_results.csv"
    print(f"\n== {run_dir.name} ==")
    if not fp.exists():
        print("Missing pipeline_results.csv")
        continue

    rows = list(csv.DictReader(open(fp, "r", encoding="utf-8")))
    print("rows:", len(rows))

    run_compiled = 0
    run_buckets = Counter()

    for row in rows:
        total_rows += 1
        comp = str(row.get("compiled","")).strip().lower() in {"1","true","yes"}
        if comp:
            compiled_rows += 1
            run_compiled += 1

        b = (row.get("error_bucket") or "").strip()
        if b:
            bucket_counter[b] += 1
            run_buckets[b] += 1

        # print first-line diagnostics per row
        print(
            "slide:", row.get("slide_id"),
            "| compiled:", row.get("compiled"),
            "| bucket:", row.get("error_bucket"),
            "| gate:", (row.get("gate_errors","") or "")[:90],
            "| err:", (row.get("stderr_or_error","") or "")[:90]
        )

    print("compiled_true:", run_compiled)
    print("run_buckets:", dict(run_buckets) if run_buckets else "{}")

compile_rate = (compiled_rows / total_rows) if total_rows > 0 else 0.0

print("\n=== 3D-LITE Smoke Summary ===")
print("total_rows:", total_rows)
print("compiled_rows:", compiled_rows)
print("compile_rate:", round(compile_rate, 3))
print("all_error_buckets:", dict(bucket_counter))

# go/no-go
if compile_rate >= 0.35:
    print("✅ Proceed to Step 7 (full 3D-LITE run).")
else:
    print("⚠️ Do NOT proceed yet; tune prompts/settings first.")
```

Build DPO

```python
# ===== Step 11 FIXED (Bplus 30x2): recover prompt/code from files =====
import json, csv, random
from pathlib import Path

random.seed(42)

DATASET_ROOT = Path("/content/drive/MyDrive/papertok_dataset_cs229")
PAIR_POOL = DATASET_ROOT / "dpo_pair_pool_auto_bplus_30x2.jsonl"
LABELS_CSV = DATASET_ROOT / "dpo_labels_filled_template_bplus_30x2.csv"
RUNS_ROOT = DATASET_ROOT / "stage2_runs_bplus_30x2_retry"

TRAIN_OUT = DATASET_ROOT / "dpo_train_bplus_30x2.jsonl"
VAL_OUT = DATASET_ROOT / "dpo_val_bplus_30x2.jsonl"

def load_jsonl(path):
    out = []
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            if line.strip():
                out.append(json.loads(line))
    return out

def load_labels(path):
    out = {}
    with open(path, "r", encoding="utf-8") as f:
        r = csv.DictReader(f)
        for row in r:
            out[row["pair_id"]] = (row.get("label", "").strip().lower(), row.get("notes", ""))
    return out

def read_prompt(run_id, slide_id):
    sj = RUNS_ROOT / run_id / "outputs" / "slides.json"
    if not sj.exists():
        return ""
    try:
        data = json.loads(sj.read_text(encoding="utf-8"))
    except Exception:
        return ""
    slides = data if isinstance(data, list) else data.get("slides", [])
    for s in slides:
        if str(s.get("id", "")) == slide_id:
            return (
                "Generate Manim code for this concept animation.\n\n"
                f"Slide ID: {slide_id}\n"
                f"Voice: {s.get('voice','')}\n"
                f"Goal: {s.get('goal','')}\n"
                f"Remarks: {s.get('remarks','')}\n"
                f"Formula: {s.get('formula','')}\n"
            ).strip()
    return ""

def read_code(run_id, slide_id):
    code_dir = RUNS_ROOT / run_id / "outputs" / "slides"
    if not code_dir.exists():
        return ""
    files = sorted(code_dir.glob(f"{slide_id}*.py"))
    if not files:
        return ""
    try:
        return files[-1].read_text(encoding="utf-8").strip()
    except Exception:
        return ""

pairs = {p["pair_id"]: p for p in load_jsonl(PAIR_POOL)}
labels = load_labels(LABELS_CSV)

examples = []
stats = {"keep_auto":0, "flip":0, "skip":0, "missing_label":0, "unusable":0}

for pid, p in pairs.items():
    if pid not in labels:
        stats["missing_label"] += 1
        continue

    label, _ = labels[pid]
    if label in {"", "skip"}:
        stats["skip"] += 1
        continue
    if label not in {"keep_auto", "flip"}:
        stats["skip"] += 1
        continue

    slide_id = p["slide_id"]
    c_run = p["chosen_candidate"]
    r_run = p["rejected_candidate"]

    prompt = read_prompt(c_run, slide_id) or read_prompt(r_run, slide_id)
    chosen = read_code(c_run, slide_id)
    rejected = read_code(r_run, slide_id)

    if label == "flip":
        chosen, rejected = rejected, chosen
        stats["flip"] += 1
    else:
        stats["keep_auto"] += 1

    if not prompt or not chosen or not rejected:
        stats["unusable"] += 1
        continue

    examples.append({
        "id": pid,
        "paper_id": p.get("paper_id", ""),
        "slide_id": slide_id,
        "prompt": prompt,
        "chosen": chosen,
        "rejected": rejected,
    })

random.shuffle(examples)
n = len(examples)
n_val = max(1, int(0.1 * n)) if n >= 10 else (1 if n >= 2 else 0)
val = examples[:n_val]
train = examples[n_val:]

with open(TRAIN_OUT, "w", encoding="utf-8") as f:
    for ex in train:
        f.write(json.dumps(ex, ensure_ascii=False) + "\n")
with open(VAL_OUT, "w", encoding="utf-8") as f:
    for ex in val:
        f.write(json.dumps(ex, ensure_ascii=False) + "\n")

print("Total labeled usable:", n)
print("keep_auto:", stats["keep_auto"], "| flip:", stats["flip"], "| skip:", stats["skip"], "| missing_label:", stats["missing_label"], "| unusable:", stats["unusable"])
print("Train:", len(train), "->", TRAIN_OUT)
print("Val:", len(val), "->", VAL_OUT)
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
from peft import PeftModel

MODEL_NAME = "Qwen/Qwen2.5-Coder-3B-Instruct"
ADAPTER_DIR = "/content/drive/MyDrive/papertok_dataset_cs229/dpo_pilot_no_bnb"

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, use_fast=True)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

base_model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    torch_dtype=torch.float16,
    device_map="auto",
    trust_remote_code=True,
)
base_model.eval()

adapt_model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    torch_dtype=torch.float16,
    device_map="auto",
    trust_remote_code=True,
)
adapt_model = PeftModel.from_pretrained(adapt_model, ADAPTER_DIR)
adapt_model.eval()

def generate_text(model, prompt, max_new_tokens=220):
    chat = [
        {"role": "system", "content": "You generate clean, compilable Manim Python code only. No explanations."},
        {"role": "user", "content": prompt},
    ]
    text = tokenizer.apply_chat_template(chat, tokenize=False, add_generation_prompt=True)
    inputs = tokenizer(text, return_tensors="pt").to(model.device)
    with torch.no_grad():
        out = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=False,
            temperature=0.0,
            pad_token_id=tokenizer.eos_token_id,
        )
    gen = tokenizer.decode(out[0][inputs["input_ids"].shape[1]:], skip_special_tokens=True)
    return gen

comparisons = []
for i, s in enumerate(samples, 1):
    p = s["prompt"]
    base_out = generate_text(base_model, p)
    dpo_out = generate_text(adapt_model, p)
    comparisons.append({
        "id": s.get("id"),
        "slide_id": s.get("slide_id"),
        "prompt": p,
        "base_output": base_out,
        "dpo_output": dpo_out,
    })
    print(f"\n===== SAMPLE {i}: {s.get('id')} / {s.get('slide_id')} =====")
    print("BASE (first 500 chars):\n", base_out[:500])
    print("\nDPO  (first 500 chars):\n", dpo_out[:500])
```

```python
import json
from pathlib import Path
from statistics import mean

DATA_ROOT = Path("/content/drive/MyDrive/papertok_dataset_cs229")
REPORT_PATH = DATA_ROOT / "dpo_pilot_eval_report.json"

report = json.loads(REPORT_PATH.read_text(encoding="utf-8"))
samples = report.get("samples", [])

if len(samples) == 0:
    raise ValueError("No samples in report.")

def safe_int(x, d=0):
    try:
        return int(x)
    except:
        return d

def total_score(d):
    return safe_int(d.get("concept_fidelity", 0)) + \
           safe_int(d.get("visual_clarity", 0)) + \
           safe_int(d.get("animation_coherence", 0)) + \
           safe_int(d.get("code_simplicity_stability", 0))

n = len(samples)

base_compile = [1 if s.get("base_compiles", False) else 0 for s in samples]
dpo_compile = [1 if s.get("dpo_compiles", False) else 0 for s in samples]

base_totals = [total_score(s.get("base_scores", {})) for s in samples]
dpo_totals = [total_score(s.get("dpo_scores", {})) for s in samples]
deltas = [d - b for b, d in zip(base_totals, dpo_totals)]

winners = [str(s.get("human_winner", "")).lower() for s in samples]
judged = [w for w in winners if w in {"base", "dpo", "tie", "both_bad"}]

dpo_wins = sum(1 for w in judged if w == "dpo")
base_wins = sum(1 for w in judged if w == "base")
ties = sum(1 for w in judged if w == "tie")
both_bad = sum(1 for w in judged if w == "both_bad")
valid_comp = sum(1 for w in judged if w in {"base", "dpo", "tie"})  # exclude both_bad

base_gate = sum(1 for s in samples if str(s.get("base_error_bucket","")) == "gate_fail")
dpo_gate = sum(1 for s in samples if str(s.get("dpo_error_bucket","")) == "gate_fail")
base_manim = sum(1 for s in samples if str(s.get("base_error_bucket","")) == "manim_error")
dpo_manim = sum(1 for s in samples if str(s.get("dpo_error_bucket","")) == "manim_error")

summary = {
    "num_samples": n,
    "compile_rate_base": round(sum(base_compile) / n, 3),
    "compile_rate_dpo": round(sum(dpo_compile) / n, 3),
    "compile_rate_delta": round((sum(dpo_compile) - sum(base_compile)) / n, 3),

    "avg_score_base": round(mean(base_totals), 3),
    "avg_score_dpo": round(mean(dpo_totals), 3),
    "avg_score_delta": round(mean(deltas), 3),

    "win_rate_dpo_all_judged": round(dpo_wins / len(judged), 3) if judged else None,
    "win_rate_dpo_excl_both_bad": round(dpo_wins / valid_comp, 3) if valid_comp else None,

    "dpo_wins": dpo_wins,
    "base_wins": base_wins,
    "ties": ties,
    "both_bad": both_bad,
    "both_bad_rate": round(both_bad / len(judged), 3) if judged else None,

    "gate_fail_rate_base": round(base_gate / n, 3),
    "gate_fail_rate_dpo": round(dpo_gate / n, 3),
    "manim_error_rate_base": round(base_manim / n, 3),
    "manim_error_rate_dpo": round(dpo_manim / n, 3),
}

print(json.dumps(summary, indent=2))

# Optional success flags for pilot
success = {
    "compile_not_worse": summary["compile_rate_delta"] >= 0,
    "dpo_win_rate_ge_055_excl_both_bad": (summary["win_rate_dpo_excl_both_bad"] is not None and summary["win_rate_dpo_excl_both_bad"] >= 0.55),
    "avg_score_delta_positive": summary["avg_score_delta"] > 0,
}
print("\nPilot success checks:")
print(json.dumps(success, indent=2))
```
